The dataset contains the following columns:

Country/Region: The name of the country or region.

Continent: The continent to which the country belongs.

Population: The population of the country or region.

TotalCases: The total number of confirmed COVID-19 cases.

NewCases: The number of new COVID-19 cases reported recently.

TotalDeaths: The total number of deaths attributed to COVID-19.

NewDeaths: The number of new deaths reported recently.

TotalRecovered: The total number of recoveries from COVID-19.

NewRecovered: The number of new recoveries reported recently.

ActiveCases: The number of currently active COVID-19 cases.

Serious,Critical: The number of cases in serious or critical condition.

Tot Cases/1M pop: The total number of cases per million people.

Deaths/1M pop: The total number of deaths per million people.

TotalTests: The total number of COVID-19 tests conducted.

Tests/1M pop: The total number of tests conducted per million people.

WHO Region: The World Health Organization (WHO) region classification.

```python
df.info()
```

NewCases ,  NewDeaths , NewRecovered  : drop this columns 

```python
df = df.drop(columns = ['NewCases','NewDeaths','NewRecovered'] )
```

```python
df.info()
```

Drop Serious,Critical 

```python
df = df.drop(columns=['Serious,Critical']) 
```

Fill  WHO Region Missing values with mode 

```python
df['WHO Region'] = df['WHO Region'].fillna(df['WHO Region'].mode()[0])
```

Fill  Deaths/1M pop Missing values with mean 

```python
df['Deaths/1M pop'] = df['Deaths/1M pop'].fillna(df['Deaths/1M pop'].mean())
```

fill TotalDeaths  missing values with mean 

```python
df['TotalDeaths'] = df['TotalDeaths'].fillna(df['TotalDeaths'].mean())
```

fill TotalTests with mean 

```python
df['TotalTests'] = df['TotalTests'].fillna(df['TotalTests'].mean())
```

fill Tests/1M pop with mean 

```python

df['Tests/1M pop'] = df['Tests/1M pop'].fillna(df['Tests/1M pop'].mean())
```

```python
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
# Select only numerical columns for the correlation matrix
numerical_titanic = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr = numerical_titanic.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```
